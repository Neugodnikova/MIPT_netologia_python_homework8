## Задание 1 ##

Напишите функцию, которая принимает на вход строку и проверяет, является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:
``` pyton
car_id = 'А222BС96'
Результат: Номер А222BС валиден. Регион: 96

car_id = 'АБ22ВВ193'
Результат: Номер не валиден
```


````python
import re

def car_number(text):
    """
    Проверяет, являеются ли данные в строке валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры).
    Если номер валиден, то функция возвращает отдельно номер и регион.
    
    """
    car_id=re.findall('([АВЕКМНОРСТУХавекмнорстухABEKMHOPCTYXabekmhopctyx]\s*\d{3}\s*[АВЕКМНОРСТУХавекмнорстухABEKMHOPCTYXabekmhopctyx]{2}\s*\d{2,3})|([АВЕКМНОРСТУХавекмнорстухABEKMHOPCTYXabekmhopctyx]{2}\s*\d{3}\s*\d{2,3})', text)  
    a=['Номер ', text[:6],' валиден. Регион: ', text[-2:]]
    return print("".join(a) if car_id else 'Номер не валиден')
````

````python
text='А222BС96'
car_number(text)
````

````python
import re
s = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'
print( re.sub(r'\b([^\W\d_]+)(\s+\1)+\b', r'\1', re.sub(r'\W+', ' ', s).strip(), flags=re.I) )
````
